## Проверка гипотез

### 1 Гипотеза 

Более 3 бракованных листов на партию выходит значимо чаще для стали марки А, чем для стали марки B.

Нулевая гипотеза (H0): частота бракованных листов (более 3) стали марки A и B 
одинакова.  
Альтернативная гипотеза (Ha): частота бракованных листов стали марки A больше, чем частота бракованных листов стали марки B.  

```python
# results = вызов метода для проверки гипотезы

steel_a = df[df['steel_grade'] == 'A']
steel_b = df[df['steel_grade'] == 'B']
```

```python
total_cost_a = steel_a[steel_a['defective_sheets'] > 3]['defective_sheets']
total_cost_b = steel_b[steel_b['defective_sheets'] > 3]['defective_sheets']
```

```python
results = st.ttest_ind(total_cost_a, total_cost_b, equal_var = False)
```

```python
alpha = .05 # alpha = задайте значение уровня значимости

print(results.pvalue)# вывод значения p-value на экран 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")# условный оператор с выводом строки с ответом
```

### 2 Гипотеза

2)При скоростях прокатки более 4 м/с свыше 3 бракованных листов стали на партию выходит значимо чаще, чем при меньших скоростях прокатки. 

Нулевая гипотеза (H0): частота бракованных листов стали более 3 бракованных листов на партию одинакова для скоростей прокатки менее или равной 4 м/с.

Альтернативная гипотеза (Ha): частота бракованных листов стали более 3 бракованных листов на партию выше для скоростей прокатки более 4 м/с.

```python
# results = вызов метода для проверк гипотезы
total_cost_a = steel_a.query("`rolling_speed` <= 4 and `defective_sheets` >3")['defective_sheets']#скоростях прокатки более 4 м/с свыше 3 бракованных листов стали для марки А
total_cost_b= steel_b.query("`rolling_speed` <= 4 and `defective_sheets` >3")['defective_sheets']#скоростях прокатки более 4 м/с свыше 3 бракованных листов стали для марки B
```

```python
results = st.ttest_ind(total_cost_a, total_cost_b, equal_var = False)
```

```python
results = st.ttest_ind(total_cost_a, total_cost_b, equal_var = False)
alpha = .05 # alpha = задайте значение уровня значимости

print(results.pvalue)# вывод значения p-value на экран 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")# условный оператор с выводом строки с ответом
```

### Немного теории

В соответствии с технологией процесса обжига для лучшего использования тепла машина разделена на технологические зоны, перекрытые сверху специальными секциями горна. Тепловой режим в каждой секции устанавливают, как правило, независимо от режима других секций. Газы из каждой зоны отсасываются отдельными дымососами. Обычно конвейерная машина состоит из следующих зон: сушки (первая камера), подогрева(вторая камера), обжига (третья камера), рекуперации(четвертая камера) и охлаждения(пятая камера).      
Влияние изменения высоты слоя тем больше, чем выше абсолютная температура газов, достигаемая по мере перемещения окатышей от зоны подогрева до зоны рекуперации.   
Таким образом, для получения максимальной прочности окатышей необходимо ограничить высоту слоя величиной 220–230 мм. (информация взята из открытых источников.
```python
# функция для подсчёта границ с учетом 1.5 размаха
def quartile_range(data, column):
    q1 = data[column].quantile(0.05)
    q3 = data[column].quantile(0.95)
    iqr = q3 - q1
    dfq = data.loc[(data[column] < q3 + 1.5*iqr) & (data[column] > q1 - 1.5*iqr), column]
    return dfq
list_features = ['t_data_1_1', 't_data_1_2', 't_data_1_3', 't_data_2_1', 't_data_2_2',
       't_data_2_3', 't_data_3_1', 't_data_3_2', 't_data_3_3', 't_data_4_1',
       't_data_4_2', 't_data_4_3', 't_data_5_1', 't_data_5_2', 't_data_5_3', 'ah_data']# столбцы с выбросами
for col in list_features:
    data_train.loc[:, col] = quartile_range(data_train, col);
```
