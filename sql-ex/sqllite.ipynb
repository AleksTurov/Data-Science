```python
sql = '''
WITH daily_sales AS (
    SELECT customer_id, date(sale_date) AS sale_date
    FROM sales
    WHERE customer_id IS NOT NULL
    GROUP BY customer_id, date(sale_date)
),

-- streaks: здесь мы генерируем идентификаторы для каждой серии покупок. Для этого мы вычитаем номер строки в окне (сгруппированном по customer_id и отсортированном по sale_date) из sale_date. 
-- В результате, одинаковые идентификаторы будут присвоены последовательным датам покупок.

streaks AS (
    SELECT customer_id, sale_date,
           date(sale_date, '-' || (ROW_NUMBER() OVER (PARTITION BY customer_id ORDER BY sale_date)) || ' day') AS streak_id
    FROM daily_sales
),
-- streak_lengths: на этом этапе мы подсчитываем длину каждой серии покупок и даты начала и конца серии. Мы также фильтруем результаты, чтобы оставить только серии длиной 3 и более дней.

streak_lengths AS (
    SELECT customer_id, streak_id, COUNT(*) AS streak_length, MIN(sale_date) AS min_date, MAX(sale_date) AS max_date
    FROM streaks
    GROUP BY customer_id, streak_id
    HAVING streak_length >= 3
),

-- streak_sales: в этом подзапросе мы вычисляем сумму и количество покупок для каждой серии покупок каждого клиента, объединяя таблицу sales с streak_lengths по customer_id.
-- Мы суммируем только те покупки, которые были сделаны во время соответствующей серии покупок.

streak_sales AS (
    SELECT sl.customer_id, sl.streak_id, sl.streak_length, sl.min_date, sl.max_date, 
    SUM(s.amount_rur) AS total_sales, COUNT(s.amount_rur) AS count_sales
    FROM sales AS s
    JOIN streak_lengths AS sl ON s.customer_id = sl.customer_id
    WHERE date(s.sale_date) BETWEEN sl.min_date AND sl.max_date
    GROUP BY s.customer_id, sl.streak_id
)

--Основной запрос: Мы объединяем таблицу streak_sales с подзапросом, который содержит максимальные длины серий для каждого клиента, и фильтруем результаты по максимальной длине серии. 
-- В результате получаем таблицу с колонками customer_id, максимальной длиной серии покупок, количеством покупок в эти даты, минимальной датой посещения в промежутке, максимальной датой посещения 
-- в промежутке и суммой покупок за эти дни. Затем мы сортируем результаты по customer_id.


SELECT
    ss.customer_id,
    ss.streak_length AS max_streak_length,
    ss.count_sales,
    ss.min_date,
    ss.max_date,
    ss.total_sales
FROM streak_sales AS ss
JOIN (
    SELECT customer_id, MAX(streak_length) AS max_streak_length
    FROM streak_lengths
    GROUP BY customer_id
) AS ms ON ss.customer_id = ms.customer_id AND ss.streak_length = ms.max_streak_length
ORDER BY ss.customer_id;
'''
```
